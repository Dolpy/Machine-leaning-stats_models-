```python
#one should be careful of the Dummy Variable Trap. so what is Dummy variable trap ?
#The Dummy Variable trap is a scenario in which the independent variables are multicollinear-a scenario in which two 
#or more variables are highly correlated; in simple terms one variable can be predicted from the others.
#The solution to the dummy variable trap is to drop one of the categorical variables 
```
